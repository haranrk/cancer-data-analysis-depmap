```python
sample_names = list(h_3.index)
sample_types = list(set([a.split('_')[1] for a in sample_names]))
```

```python
h_3.columns
```

```python
cluster_3 = {}
for c in h_3.columns:
    cluster_3[c] = []
    for r in h_3.index:
        if h_3[c][r]==1:
            cluster_3[c].append(r.split('_')[1])

pie_data_3 = {}
for x in cluster_3:
    pie_data_3[x] = []
    for y in sample_types:
        pie_data_3[x].append(cluster_3[x].count(y))

for z in pie_data_3:
    print(z)
    plt.figure(figsize=(10,8))
    plt.pie(pie_data_3[z], labels=sample_types, autopct='%1.1f')
    plt.axis('equal')
    plt.show()
```
